```python
columns = {'ignition_type': [], 'Fuel_Type': [],'body_type': [],'kms_driven': [],'transmission_type': [],'owner_No': [],'owner': [],
           'manufacturer':[],'model': [],'Manufacture_year':[],'variantName':[],'centralVariantId':[],'price':[],'priceActual': [],
           'priceSaving': [],'priceFixedText': []}
for dic in details_list:
    columns['ignition_type'].append(dic['it'])
    columns['Fuel_Type'].append(dic['ft'])
    columns['body_type'].append(dic['bt'])
    columns['kms_driven'].append(dic['km'])
    columns['transmission_type'].append(dic['transmission'])
    columns['owner_No'].append(dic['ownerNo'])
    columns['owner'].append(dic['owner'])
    columns['manufacturer'].append(dic['oem'])
    columns['model'].append(dic['model'])
    columns['Manufacture_year'].append(dic['modelYear'])
    columns['variantName'].append(dic['variantName'])
    columns['centralVariantId'].append(dic['centralVariantId'])
    columns['price'].append(dic['price'])
    columns['priceActual'].append(dic['priceActual'])
    columns['priceSaving'].append(dic['priceSaving'])
    columns['priceFixedText'].append(dic['priceFixedText'])

details_df = pd.DataFrame(columns)
details_df.head()
```

# Car overview

```python
import ast
import pandas as pd

def car_overview(d, parent_key='', sep='_'):
    items = []
    for key, value in d.items():
        new_key = f'{parent_key}{sep}{key}' if parent_key else key
        if isinstance(value, dict):
            items.extend(car_overview(value, new_key, sep).items()) 
        else:
            items.append((new_key, value))  
    return dict(items)

overview_data = []
for index, row in df.iterrows():
    try:
        unstructured_data = ast.literal_eval(row['new_car_overview'])
        overview_row = car_overview(unstructured_data)  
        overview_data.append(overview_row)
    
    except Exception as e:
        print(f"Error at index {index}: {e}") 

overview_df = pd.DataFrame(overview_data)
overview_df.head()
```

```python
def overview_dict(row):
    if isinstance(row, list):
        return {d['key']: d['value'] for d in row}
    return{}
df['new_car_overview'] = overview_df['top'].apply(overview_dict)
overview_df = pd.DataFrame(df['new_car_overview'].tolist())
overview_df.head()
```

# Car feature

```python
import ast
import pandas as pd
def feature_dict(row):
    feature_data = ast.literal_eval(row)
    features = [item['value'] for item in feature_data.get('top', [])]
    return features

df['features'] = df['new_car_feature'].apply(feature_dict)
features_df = df.explode('features')[['features']]
features_df.reset_index(drop=True, inplace=True)
features_df.head()
```

# Car Specifications

```python
import ast
import pandas as pd

def car_spec(d, parent_key='', sep='_'):
    items = []
    for key, value in d.items():
        new_key = f'{parent_key}{sep}{key}' if parent_key else key
        if isinstance(value, dict):
            items.extend(car_spec(value, new_key, sep).items()) 
        else:
            items.append((new_key, value))  
    return dict(items)

spec_data = []
for index, row in df.iterrows():
    try:
        unstructured_data = ast.literal_eval(row['new_car_specs'])
        spec_row = car_spec(unstructured_data)  
        spec_data.append(spec_row)
    
    except Exception as e:
        print(f"Error at index {index}: {e}") 

spec_df = pd.DataFrame(spec_data)
spec_df.head()
```

```python
def spec_dict(row):
    if isinstance(row,list):
        return {d['key']:d['value'] for d in row}
    return{}
df['new_car_specs'] = spec_df['top'].apply(spec_dict)
spec_df = pd.DataFrame(df['new_car_specs'].tolist())
spec_df
```

# Converting string to numerical

```python
import pandas as pd
import re
def extract_numbers(text):
    numbers = re.findall(r'\d+\.?\d*', str(text))  
    return numbers[0] if numbers else None 
       
car_df['Mileage'] = car_df['Mileage'].apply(extract_numbers).astype(float)
car_df['Engine'] =car_df['Engine'].apply(extract_numbers).astype(float)
car_df['Max Power'] = car_df['Max Power'].apply(extract_numbers).astype(float)
car_df['Torque'] = car_df['Torque'].apply(extract_numbers).astype(float)
```

```python
def price_convert(price):
    if pd.isna(price):
        return None    
    price = str(price).strip()      
    if 'Lakh' in price:
        return float(re.sub(r'[^\d.]', '', price)) * 100000
    elif 'Cr' in price:
        return float(re.sub(r'[^\d.]', '', price)) * 10000000
    else:
        cleaned_price = re.sub(r'[^\d.]', '', price)
        return float(cleaned_price) if cleaned_price else None
car_df['price'] = car_df['price'].apply(price_convert)
```

# Linear Regression Model

```python
x = car_df.drop(['price'],axis=1)
y = car_df['price']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


lr=LinearRegression()
lr.fit(x_train,y_train)

train_pred = lr.predict(x_train)
test_pred = lr.predict(x_test)
```

```python
print('*********Train data*******')
print('R2 score:',metrics.r2_score(y_train, train_pred ))
print("MAE :",metrics.mean_absolute_error(y_train,train_pred ))
print("MSE :",metrics.mean_squared_error(y_train,train_pred ))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_train,train_pred )))

print('**********Test data********')
print('R2 score:',metrics.r2_score(y_test,test_pred))
print("MAE :",metrics.mean_absolute_error(y_test,test_pred))
print("MSE :",metrics.mean_squared_error(y_test,test_pred))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test,test_pred)))
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV

xgbr = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

param_grid = {
    'n_estimators': [100, 300, 500, 700],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 1, 5],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 1.5, 2]
}

random_model = RandomizedSearchCV(
    estimator=xgbr,
    param_distributions=param_grid,
    n_iter=50,
    scoring='r2',
    cv=5,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

random_model.fit(x_train, y_train)
best_model = random_model.best_estimator_
train_pred = best_model.predict(x_train)
test_pred = best_model.predict(x_test)

print('*********Train data*******')
print('R2 score:',metrics.r2_score(y_train, train_pred ))
print("MAE :",metrics.mean_absolute_error(y_train,train_pred ))
print("MSE :",metrics.mean_squared_error(y_train,train_pred ))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_train,train_pred )))

print('**********Test data********')
print('R2 score:',metrics.r2_score(y_test,test_pred))
print("MAE :",metrics.mean_absolute_error(y_test,test_pred))
print("MSE :",metrics.mean_squared_error(y_test,test_pred))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test,test_pred)))
```
